```
import matplotlib.pyplot as plt
import numpy as np
```

```
with open("record_static.txt") as f:
    lines = [line.strip() for line in f]
arr = np.asarray(lines, dtype=np.float32)
angle = arr

plt.plot(angle)
```

```
s = np.std(angle)

print("The standar deviation is: {}".format(s))
```

```
import math
ua = s / math.sqrt(np.size(angle))
print("The uncertainty type A is: {}".format(ua))
```
